```python

from diffusers import StableDiffusionPipeline, DDIMScheduler
import torch
import requests
from PIL import Image
from io import BytesIO
import numpy as np
from fastai.vision.all import *
from fastai.vision.all import *
from matplotlib import pyplot as plt

has_cuda = torch.cuda.is_available()
device = torch.device('cpu' if not has_cuda else 'cuda')
```

```python
model_id_or_path = "CompVis/stable-diffusion-v1-4"
scheduler = DDIMScheduler.from_config(model_id_or_path, subfolder="scheduler")
pipe = StableDiffusionPipeline.from_pretrained(
    model_id_or_path,
    safety_checker=None,
    use_auth_token=True,
    scheduler = scheduler,
).to(device)
```

```python
def preprocess_image(image):
    image = np.array(image.convert("RGB"))
    image = image[None].transpose(0, 3, 1, 2)
    image = torch.from_numpy(image).to(dtype=torch.float32) / 127.5 - 1.0
    return image
```

```python
@torch.no_grad()
def ddim_inversion(latents, encoder_hidden_states, noise_scheduler, unet):
    next_latents = latents
    all_latents = [latents.detach().cpu()]

    # since we are adding noise to the image, we reverse the timesteps list to start at t=0
    reverse_timestep_list = reversed(noise_scheduler.timesteps)

    for i in range(len(reverse_timestep_list)-1):
        timestep = reverse_timestep_list[i]
        next_timestep = reverse_timestep_list[i+1]
        latent_model_input = noise_scheduler.scale_model_input(next_latents, timestep)
        noise_pred = unet(latent_model_input, timestep, encoder_hidden_states).sample

        alpha_prod_t =  noise_scheduler.alphas_cumprod[timestep]
        alpha_prod_t_next = noise_scheduler.alphas_cumprod[next_timestep]
        beta_prod_t = 1 - alpha_prod_t
        beta_prod_t_next = 1 - alpha_prod_t_next

        f = (next_latents - beta_prod_t ** 0.5 * noise_pred) / (alpha_prod_t ** 0.5)
        next_latents = alpha_prod_t_next ** 0.5 * f + beta_prod_t_next ** 0.5 * noise_pred
        all_latents.append(next_latents.detach().cpu())


    return all_latents

def null_text_inversion(
    pipe,
    all_latents,
    prompt,
    num_opt_steps=8,
    lr=0.01,
    tol=1e-5,
    guidance_scale=7.5,
    eta: float = 0.0,
    generator=None
):
    # null text embeddings(texte vide)
    null_text_prompt = ""
    null_text_input = pipe.tokenizer(
        null_text_prompt,
        padding="max_length",
        max_length=pipe.tokenizer.model_max_length,
        truncation=True,
        return_tensors="pt",
    )

    # empty prompt
    with torch.no_grad():
        null_text_embeddings_init = pipe.text_encoder(
            null_text_input.input_ids.to(pipe.device)
        )[0]


    null_text_embeddings = torch.nn.Parameter(null_text_embeddings_init.clone(), requires_grad=True)

    #  l’optimiseur
    optimizer = torch.optim.Adam([null_text_embeddings], lr=lr)

    #prompt d’entrée
    prompt_input = pipe.tokenizer(
        [prompt] if isinstance(prompt, str) else prompt,
        padding="max_length",
        max_length=pipe.tokenizer.model_max_length,
        truncation=True,
        return_tensors="pt",
    )

    with torch.no_grad():
        text_embeddings = pipe.text_encoder(prompt_input.input_ids.to(pipe.device))[0]


    extra_step_kwargs = pipe.prepare_extra_step_kwargs(generator, eta)

    all_null_texts = []
    latents = all_latents[-1].to(pipe.device)

    for timestep, prev_latents in pipe.progress_bar(zip(pipe.scheduler.timesteps, reversed(all_latents[:-1]))):
        prev_latents = prev_latents.to(pipe.device).detach()

        # input du modèle
        latent_model_input = pipe.scheduler.scale_model_input(latents, timestep).detach()

        #Bruit prédit avec prompt réel
        with torch.no_grad():
            noise_pred_text = pipe.unet(
                latent_model_input, timestep, encoder_hidden_states=text_embeddings
            ).sample.detach()

        for _ in range(num_opt_steps):
            #Bruit prédit avec texte nul
            noise_pred_uncond = pipe.unet(
                latent_model_input, timestep, encoder_hidden_states=null_text_embeddings
            ).sample

            #Guidance
            noise_pred = noise_pred_uncond + guidance_scale * (noise_pred_text - noise_pred_uncond)

            # Estimation des latents précédents
            prev_latents_pred = pipe.scheduler.step(
                noise_pred, timestep, latents, **extra_step_kwargs
            ).prev_sample

            #  perte
            loss = torch.nn.functional.mse_loss(prev_latents_pred, prev_latents).mean()

            #Backprop
            loss.backward()
            optimizer.step()
            optimizer.zero_grad()

        all_null_texts.append(null_text_embeddings.detach().cpu())
        latents = prev_latents_pred.detach()

    return all_latents[-1], all_null_texts
```

```python
@torch.no_grad()
def reconstruct(pipe, latents, prompt, null_text_embeddings, guidance_scale=7.5, generator=None, eta=0.0):

    prompt_input = pipe.tokenizer(
        [prompt] if isinstance(prompt, str) else prompt,
        padding="max_length",
        max_length=pipe.tokenizer.model_max_length,
        truncation=True,
        return_tensors="pt",
    )

    text_embeddings = pipe.text_encoder(prompt_input.input_ids.to(pipe.device))[0]

    extra_step_kwargs = pipe.prepare_extra_step_kwargs(generator, eta)

    latents = latents.to(pipe.device)

    for i, (t, null_text_t) in enumerate(pipe.progress_bar(zip(pipe.scheduler.timesteps, null_text_embeddings))):

        latent_model_input = torch.cat([latents] * 2)
        latent_model_input = pipe.scheduler.scale_model_input(latent_model_input, t)

        input_embedding = torch.cat([null_text_t.to(pipe.device), text_embeddings])

        #Prédiction du bruit
        noise_pred = pipe.unet(latent_model_input, t, encoder_hidden_states=input_embedding).sample

        #Guidance
        noise_pred_uncond, noise_pred_text = noise_pred.chunk(2)
        noise_pred = noise_pred_uncond + guidance_scale * (noise_pred_text - noise_pred_uncond)

        latents = pipe.scheduler.step(noise_pred, t, latents, **extra_step_kwargs).prev_sample

    image = pipe.decode_latents(latents)
    return image
```

```python
image = Image.open("/content/leo_no_glasses.jpg")
image = image.resize((512, 512))

pp_image = preprocess_image(image)

generator = torch.Generator(device=device)
latent = pipe.vae.encode(pp_image.to(device)).latent_dist.sample(generator=generator) *  0.18215

input_prompt = ["Frontal portrait of a young man with black hair and glasses"]

text_embeddings = pipe.text_encoder(
    pipe.tokenizer(input_prompt, return_tensors="pt", padding=True).input_ids.to(device)
)[0]

num_inference_steps = 50
pipe.scheduler.set_timesteps(num_inference_steps, device=device)
all_latents = ddim_inversion(latent, text_embeddings, pipe.scheduler, pipe.unet)

z_T, all_null_texts = null_text_inversion(pipe, all_latents, input_prompt, num_opt_steps=int(num_inference_steps/5))

recon_img = reconstruct(pipe, z_T, input_prompt, all_null_texts, guidance_scale=1)

show_image(pp_image[0].permute(1,2,0).numpy() * 0.5 + 0.5)
show_image(recon_img.squeeze())
```

```python
import lpips
import torch
import numpy as np
from PIL import Image
from skimage.metrics import peak_signal_noise_ratio as compare_psnr

# Initialiser le modèle LPIPS
loss_fn = lpips.LPIPS(net='alex') # Choisissez 'alex', 'vgg' ou 'squeeze'

def load_image(path):
  img = Image.open(path).convert('RGB').resize((256, 256))
  img_np = np.array(img).astype(np.float32) / 255.0
  img_tensor = torch.tensor(img_np).permute(2, 0, 1).unsqueeze(0) * 2 - 1 # Normalisation [-1, 1]
  return img_tensor, img_np

# Chemins vers les images
ref_path = 'original.jpg'
recon_paths = ['DDIM/10.jpg', 'DDIM/30.jpg', 'DDIM/40.jpg','DDIM/50.jpg', 'DDIM/80.jpg', 'DDIM/120.jpg', 'DDIM/300.jpg', 'NT/10.jpg', 'NT/30.jpg', 'NT/40.jpg','NT/50.jpg', 'NT/80.jpg', 'NT/120.jpg', 'NT/300.jpg', 'prompt.jpg']

# Charger l'image de référence
ref_tensor, ref_np = load_image(ref_path)

# Calculer les métriques pour chaque image reconstruite
for path in recon_paths:
  recon_tensor, recon_np = load_image(path)

  # Calcul LPIPS
  lpips_value = loss_fn(ref_tensor, recon_tensor).item()

  # Calcul PSNR
  psnr_value = compare_psnr(ref_np, recon_np, data_range=1.0)

  print(f"{path} - LPIPS: {lpips_value:.4f}, PSNR: {psnr_value:.2f} dB")
```

```python
import os
import cv2
import torch
import lpips
import numpy as np
import pandas as pd
from tqdm import tqdm
from skimage.metrics import peak_signal_noise_ratio as psnr

# Initialiser le modèle LPIPS avec le backbone 'alex'
loss_fn = lpips.LPIPS(net='alex') # Utilise le réseau AlexNet

# Définir le répertoire contenant les images
image_dir = '' # Remplacez par le chemin réel

# Initialiser une liste pour stocker les résultats
results = []

# Boucle sur les indices d'image
for i in tqdm(range(1, 23)):
  # Construire les chemins des images
  original_path = os.path.join(image_dir, f'original{i}.jpg')
  recon_nt_path = os.path.join(image_dir, f'reconstruction{i}.jpg')
  recon_ddim_path = os.path.join(image_dir, f'reconstruction{i}_ddim.jpg')

  # Vérifier l'existence des fichiers
  if not all(os.path.exists(p) for p in [original_path, recon_nt_path, recon_ddim_path]):
    print(f"Images manquantes pour l'indice {i}, saut.")
    continue

  # Charger les images avec OpenCV
  original = cv2.imread(original_path)
  recon_nt = cv2.imread(recon_nt_path)
  recon_ddim = cv2.imread(recon_ddim_path)

  # Redimensionner les images pour qu'elles aient la même taille
  h, w = original.shape[:2]
  recon_nt = cv2.resize(recon_nt, (w, h))
  recon_ddim = cv2.resize(recon_ddim, (w, h))

  # Convertir les images de BGR à RGB
  original_rgb = cv2.cvtColor(original, cv2.COLOR_BGR2RGB)
  recon_nt_rgb = cv2.cvtColor(recon_nt, cv2.COLOR_BGR2RGB)
  recon_ddim_rgb = cv2.cvtColor(recon_ddim, cv2.COLOR_BGR2RGB)

  # Normaliser les images pour LPIPS : [0,255] -> [-1,1]
  def preprocess(img):
    img = img.astype(np.float32) / 255.0 # [0,1]
    img = img * 2 - 1 # [-1,1]
    img = torch.from_numpy(img).permute(2, 0, 1).unsqueeze(0) # Shape: [1,3,H,W]
    return img

  img0 = preprocess(original_rgb)
  img1_nt = preprocess(recon_nt_rgb)
  img1_ddim = preprocess(recon_ddim_rgb)

  # Calculer LPIPS
  lpips_nt = loss_fn(img0, img1_nt).item()
  lpips_ddim = loss_fn(img0, img1_ddim).item()

  # Calculer PSNR
  psnr_nt = psnr(original_rgb, recon_nt_rgb, data_range=255)
  psnr_ddim = psnr(original_rgb, recon_ddim_rgb, data_range=255)

  # Ajouter les résultats à la liste
  results.append({
  'index': i,
  'psnr_nt': psnr_nt,
  'psnr_ddim': psnr_ddim,
  'lpips_nt': lpips_nt,
  'lpips_ddim': lpips_ddim
  })

# Convertir les résultats en DataFrame et les enregistrer dans un fichier CSV
df = pd.DataFrame(results)
df.to_csv('similarity_metrics.csv', index=False)
print("Calcul terminé. Résultats enregistrés dans 'similarity_metrics.csv'.")
```
